```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/drive/MyDrive/cyberlabs/Polynomial_train.csv")
```

```python
data = np.array(data)
#data = data[:100,:]
print(data.shape)
m , n = data.shape  ## n is the number of features and m is the no. of exapmles
n = n - 2
print(m,n)
```

```python
y = (data[:,n+1:])/100
X = (data[:,1:n+1])/100
print(X.shape,y.shape)
```

```python
X_sq = np.multiply(X,X)
print(X_sq.shape)
```

```python
X = np.concatenate((X,X_sq),axis = 1)
print(X.shape)
```

```python
z = np.ones((m,1))
print(z.shape)
```

```python
X = np.concatenate((z,X),axis = 1)
print(X.shape)
```

```python
m,n = X.shape
n = n - 1
print(m,n)
```

```python
theta = np.zeros((n+1,1))
print(theta.shape)
```

```python
for i in range(100000) :
  h = np.dot(X,theta)
  #print(h.shape)
  er = (h - y)
  delta = (np.dot(X.T,er))/m
  theta = theta - 0.0001*delta
  J = (1/(2*m))*(np.sum(np.multiply(er,er)))
print(J)
print(theta)
```
